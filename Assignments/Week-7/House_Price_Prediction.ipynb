## Data Cleaning

```python
df[['Alley','MiscFeature','Fence','PoolQC']].nunique()
```

```python
#dropping irrelevant columns and columns with maximum null values
df.drop(columns=['Id','Alley','MiscFeature','Fence','PoolQC'],axis=1,inplace=True)
```

```python
# Numerical columns: filling with mean values
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].apply(lambda x: x.fillna(x.mean()), axis=0)

# Categorical columns: filling with mode values
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]), axis=0)
df
```

## Data Preprocessing

```python
#Converting columns containing objects into numeric data
from sklearn.preprocessing import LabelEncoder

def encode_object_columns(df):
    le = LabelEncoder()
    for col in df.select_dtypes(include='object').columns:
        df[col] = le.fit_transform(df[col]) 
    return df

df = encode_object_columns(df)
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler

def scale_numeric_columns(df):
    numeric_columns = [
        'LotFrontage', 'LotArea', 'MasVnrArea', 
        'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', 
        '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'GarageYrBlt', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 
        'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 
        'MoSold'
    ]
    
    scaler = StandardScaler()
    
    df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
    return df

df = scale_numeric_columns(df)
```

## Splitting into Training & Testing Sets

```python
#Separating the features and the target variable
X=df.drop(columns=['SalePrice'])
y=df['SalePrice']
```

```python
#breaking the dataset into 70% training data and 30% testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

## Machine Learning Algorithms

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Defining the models
models = {
    'Linear Regression': LinearRegression(),
    'Lasso Regression': Lasso(),
    'Ridge Regression': Ridge(),
    'ElasticNet Regression': ElasticNet(),
    'Support Vector Machine': SVR(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(n_estimators=100)
}

# Training and evaluate the models based on R score
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} R Score: {r2}")
```
