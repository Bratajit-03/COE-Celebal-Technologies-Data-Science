## Importing Necessary Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Loading the Image

```python
image1 = cv2.imread('/content/image1.jpg')
image2 = cv2.imread('/content/image2.jpg')
```

## Converting Image to RGB

```python
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

## Histograms for each color channel

```python
# Plotting histogram for image1
colors = ('r', 'g', 'b')
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.subplot(1, 2, 2)
plt.title('Color Histogram for Image1')
plt.xlabel('Bins')
plt.ylabel('Number of Pixels')
for i, col in enumerate(colors):
    histogram = cv2.calcHist([image1], [i], None, [256], [0, 256])
    plt.plot(histogram, color=col)
    plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
# Plotting histogram for image2
colors = ('r', 'g', 'b')
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image2)
plt.subplot(1, 2, 2)
plt.title('Color Histogram for Image2')
plt.xlabel('Bins')
plt.ylabel('Number of Pixels')
for i, col in enumerate(colors):
    histogram = cv2.calcHist([image2], [i], None, [256], [0, 256])
    plt.plot(histogram, color=col)
    plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
