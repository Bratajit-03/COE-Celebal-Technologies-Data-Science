## Importing Necessary Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Loading the Image

```python
image1 = cv2.imread('/content/image1.jpg')
image2 = cv2.imread('/content/image2.jpg')
```

## Converting Image to RGB

```python
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

## Histograms for each color channel

```python
#Histogram for image1
color = ('r', 'g', 'b')
plt.figure()
plt.title('Color Histogram for Image-1')
plt.xlabel('Bins')
plt.ylabel('Number of Pixels')
for i, col in enumerate(color):
    histogram = cv2.calcHist([image1], [i], None, [256], [0, 256])
    plt.plot(histogram, color=col)
    plt.xlim([0, 256])
```

```python
#Histogram for image2
color = ('r', 'g', 'b')
plt.figure()
plt.title('Color Histogram for Image-2')
plt.xlabel('Bins')
plt.ylabel('Number of Pixels')
for i, col in enumerate(color):
    histogram = cv2.calcHist([image2], [i], None, [256], [0, 256])
    plt.plot(histogram, color=col)
    plt.xlim([0, 256])
```
